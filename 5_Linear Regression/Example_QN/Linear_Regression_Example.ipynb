```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

Importing Libraries


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import files
```

Reading Dataset Via Kaggle


```python
files.upload()
```

```python
data=pd.read_csv("salary_Data.csv")
data.head()
```

Seeing the Dimensions

```python
data.shape
```

Data Cleaning - Seeing for null values (All values present so 0)

```python
data.isnull().sum()
```

Assigning Independent Variables & Dependent Variables

```python
x=data.iloc[:, :1].values
y=data.iloc[:, 1:2].values
```

Splitting the dataset into 2

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=42,
)
```

Fit the model

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

Testing model

```python
y_pred=model.predict(x_test)
y_pred
```

Actual vals

```python
y_test
```

Almost Ok, but yet there is some error % error in (act-pred) values

Plot for Training set

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,model.predict(x_train),color="blue")
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Plot of test set

```python
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,model.predict(x_test),color="blue")
plt.title("Salary vs Experience (Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
